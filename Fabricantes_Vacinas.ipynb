```python
dados['vacina_dataAplicacao'] = pd.to_datetime(dados['vacina_dataAplicacao'], format='%Y-%m-%d')
dados_primeira_dose = dados.loc[dados['vacina_descricao_dose'] == '1ª Dose']
dados_primeira_dose.sort_values('vacina_dataAplicacao', ascending= False)
```

```python
# contar o número de doses por fabricante e por tipo de dose
dados_dose = dados.groupby(['vacina_fabricante_nome', 'vacina_descricao_dose'])['vacina_descricao_dose'].count()

# converter a série em um dataframe com colunas separadas para cada tipo de dose
dados_dose = dados_dose.unstack()

# exibir o resultado
dados_dose
```

# Tratamento dos dados

```python
# Tratamento dos dados de fabricantes das vacinas

# substituir os nomes das fabricantes que deseja agrupar por um novo nome
dados['vacina_fabricante_nome'] = dados['vacina_fabricante_nome'].replace({
    'PFIZER - PEDIÁTRICA': 'PFIZER_PEDIÁTRICA', 
    'PFIZER - PEDIÁTRICA MENOR DE 5 ANOS': 'PFIZER_PEDIÁTRICA',
    'PFIZER - PEDI?TRICA': 'PFIZER_PEDIÁTRICA',
    'Pendente Identifica??o': 'Pendente Identificação',
    'Pendente Identifica??o': 'Pendente Identificação',
    'ASTRAZENECA/FIOCRUZ': 'ASTRAZENECA',
    'PFIZER MANUFACTURING BELGIUM NV - BELGICA': 'PFIZER',
    'SINOVAC/BUTANTAN': 'SINOVAC',
    'FUNDACAO OSWALDO CRUZ': 'ASTRAZENECA'})

# agrupar as quantidades por dose, considerando o novo nome das fabricantes
dados_dose = dados.groupby(['vacina_fabricante_nome', 'vacina_descricao_dose'])['vacina_descricao_dose'].count().unstack()

dados_dose
```

```python
# agrupar as vacinas por dose
dados_dose['1ª Dose Total'] = dados_dose[['1ª Dose', '1ª Dose Revacinação', '1º Reforço']].sum(axis=1)
dados_dose['2ª Dose Total'] = dados_dose[['2ª Dose', '2ª Dose Revacinação', '2º Reforço']].sum(axis=1)

dados_dose
```

```python
dict_fab = dados_dose.to_dict(orient='dict')
dict_fab
```

```python
# Manter apenas 1ª dose e 2ª dose
# Criação dos dataframes
p_dose = pd.DataFrame({ '1ª Dose Total': {'ASTRAZENECA': 260591.0,
  'JANSSEN': 588.0,
  'PFIZER': 187791.0,
  'PFIZER_PEDIÁTRICA': 15491.0,
  'Pendente Identificação': 17198.0,
  'SINOVAC': 179211.0}})

s_dose = pd.DataFrame({'2ª Dose Total': {'ASTRAZENECA': 255379.0,
  'JANSSEN': 14103.0,
  'PFIZER': 176314.0,
  'PFIZER_PEDIÁTRICA': 12444.0,
  'Pendente Identificação': 6576.0,
  'SINOVAC': 146355.0}})

# Mesclar os dois dataframes
dose_grup = pd.merge(p_dose, s_dose, left_index=True, right_index=True)

# Ordenar os valores pelo total de doses
dose_grup = dose_grup.sort_values(by=['1ª Dose Total'], ascending=False)
dose_grup
```

```python
dose_grup.plot(kind='bar')
```
